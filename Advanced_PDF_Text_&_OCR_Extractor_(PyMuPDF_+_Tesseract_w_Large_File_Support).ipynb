````python
# Install required packages
!sudo apt-get update
!sudo apt-get install -y poppler-utils tesseract-ocr  # Install both required system packages
!pip install pytesseract pdf2image

import pytesseract
from pdf2image import convert_from_path
import os
from google.colab import drive
from PIL import Image
from IPython.display import display, Markdown, HTML
import matplotlib.pyplot as plt
import warnings

# Disable decompression bomb protection for large images
from PIL import Image
Image.MAX_IMAGE_PIXELS = None

# Suppress warnings
warnings.filterwarnings('ignore')

# Mount Google Drive (force remount if needed)
drive.mount('/content/drive', force_remount=True)

# Define directory path containing PDFs
directory_path = '/content/drive/MyDrive/Essay/'

# List of specific PDFs to process
pdf_files = ['FinalSubmission.pdf', 'WrittenAnswers.pdf']

# Verify poppler installation
!which pdftoppm  # This should show the path to pdftoppm

def extract_text_with_ocr(pdf_path):
    try:
        # Convert PDF to images with optimized settings
        images = convert_from_path(pdf_path,
                                 dpi=200,  # Reduced DPI for large pages
                                 poppler_path='/usr/bin',
                                 thread_count=4,
                                 fmt='jpeg',
                                 strict=False)

        full_text = ""

        for i, image in enumerate(images):
            temp_image_path = f"/content/temp_page_{i+1}.jpg"

            try:
                # Create a clear section for each page
                display(HTML(f"<h2 style='color:#1a5276;'>{os.path.basename(pdf_path)} - Page {i+1}</h2>"))

                # Display the original page image (downscaled if needed)
                plt.figure(figsize=(10, 10))
                plt.imshow(image)
                plt.axis('off')
                plt.title(f"Original Page {i+1}", pad=20)
                plt.show()

                # Save temporary image with quality compromise
                image.save(temp_image_path, 'JPEG', quality=85)

                # Perform OCR with timeout
                text = pytesseract.image_to_string(Image.open(temp_image_path), timeout=60)

                # Display extracted text with formatting
                display(Markdown("### Extracted Text:"))
                display(Markdown(f"```\n{text}\n```"))

                # Add to full text with page separator
                full_text += f"=== Page {i+1} ===\n{text}\n\n"

            except Exception as page_error:
                error_msg = f" Error processing page {i+1}: {str(page_error)}"
                print(error_msg)
                full_text += f"=== Page {i+1} ERROR ===\n{error_msg}\n\n"
            finally:
                # Clean up temp file if it exists
                if os.path.exists(temp_image_path):
                    os.remove(temp_image_path)

            # Add separator between pages
            display(HTML("<hr style='border:2px solid #aed6f1'>"))

        return full_text

    except Exception as e:
        print(f" Processing error for {pdf_path}: {str(e)}")
        return None

# Process specific PDFs
print(f"Processing PDFs in directory: {directory_path}")

for filename in pdf_files:
    pdf_path = os.path.join(directory_path, filename)

    if not os.path.exists(pdf_path):
        print(f"\n File not found: {filename} - Skipping...")
        continue

    print(f"\n{'='*60}")
    print(f"STARTING PROCESSING: {filename}")
    print(f"{'='*60}")

    ocr_text = extract_text_with_ocr(pdf_path)

    if ocr_text is not None:
        # Save the extracted text
        output_filename = f"OCR_extraction_{os.path.splitext(filename)[0]}.txt"
        output_path = os.path.join(directory_path, output_filename)

        with open(output_path, 'w') as f:
            f.write(ocr_text)

        print(f"\n Extraction saved to: {output_path}")
    else:
        print(f"\n Failed to process: {filename}")

    print(f"{'='*60}")
    print(f"FINISHED PROCESSING: {filename}")
    print(f"{'='*60}\n\n")
````
